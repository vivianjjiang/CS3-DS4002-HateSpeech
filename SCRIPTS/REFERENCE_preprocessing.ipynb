```
# Imports
import pandas as pd
import numpy as np

pd.set_option('display.max_colwidth', 200)
```

```
# Load the labeled dataset
data_path = 'processed/labeled_data_clean.csv'  # adjust if needed
df = pd.read_csv(data_path)

df.head()
```

```
df.info()
```

```
df['label'].value_counts()
```

```
# Map numeric labels to names (adjust mapping if needed)
label_map = {
    0: 'Hate speech',
    1: 'Offensive language',
    2: 'Neutral'
}

df['label_name'] = df['label'].map(label_map)
df[['tweet', 'label_name']].head(10)
```

```
# Example of very light text normalization
import re

def basic_clean(text):
    text = str(text).lower()
    text = re.sub(r'http\S+', 'URL', text)        # replace URLs
    text = re.sub(r'@\w+', '@USER', text)         # replace mentions
    text = re.sub(r'#[A-Za-z0-9_]+', 'HASHTAG', text)  # replace hashtags
    return text

df['tweet_clean'] = df['tweet'].apply(basic_clean)
df[['tweet', 'tweet_clean']].head()
```
